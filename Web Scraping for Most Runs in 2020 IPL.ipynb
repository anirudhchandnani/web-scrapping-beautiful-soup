```python
last_names = []
def last_name_extract():
    for last_name in soup.find_all('span' , {"class" : "top-players__last-name"}):
        last_names.append(last_name)
    return last_names
```

```python
def name_extract(your_class):
    names = []
    for name in soup.find_all('span' , {"class" : your_class}):
        names.append(name)
    return names
```
